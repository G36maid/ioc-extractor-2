### 修正MinIO上存取方式錯誤的情資
- 原本命名原則是使用pulse name
- 但有些pulse name上有 "/" ，造成儲存時會多了一個folder
- 這邊用在修改有問題的folders

```python
import pandas as pd
df = pd.read_json('test.json', lines=True)
```

```python
test3 = pd.DataFrame(data={
    'id': df[df.name.str.contains(':') == True]['id'],
    'name': df[df.name.str.contains(':') == True]['name'],
    'references': df[df.name.str.contains(':') == True]['references']
})
test3
```

```python
test2 = pd.DataFrame(data={
    'id': df[df.name.str.contains('\.') == True]['id'],
    'name': df[df.name.str.contains('\.') == True]['name'],
    'references': df[df.name.str.contains('\.') == True]['references']
})
test2
```

```python
test = pd.DataFrame(data={'id': df[df.name.str.contains('/') == True]['id'], 
    'name': df[df.name.str.contains('/') == True]['name'], 'references': df[df.name.str.contains('/') == True]['references']})
test
```

```python
combine = pd.concat([test, test2, test3], ignore_index=True)
combine.drop_duplicates("id")
combine
```

```python
from minio import Minio
from minio.error import S3Error

client = Minio(
    "192.168.70.34:9000",
    access_key="cynthiachan",
    secret_key="Mynameis375",
    secure=False
)
```

```python
for k in list(combine.id):
    for i in client.list_objects("feed-references", f'{k}/', recursive=True):
        client.remove_object("feed-references", i.object_name)
```

```python
combine.name = combine.name.str.replace('/', '_')
combine.name = combine.name.str.replace('.', '-')
combine.name = combine.name.str.replace(':', '--')
combine
```

```python
import requests, io
for row in combine.iterrows():
    idx = 0
    if len(row[1]['references']) > 0:
        for ref in row[1]['references']:
            idx+=1
            try:
                response = requests.get(ref, timeout=10)
                title = "{}-{}".format(row[1]['name'], idx)
                id = row[1]['id']
                client.put_object("feed-references", f'{id}/{title}', io.BytesIO(response.content), len(response.content))
                print("ID: {}. Successfully upload {} in Bucket 'feed-references'.".format(id, title))
            except Exception as e:
                print(e)
                pass
```
